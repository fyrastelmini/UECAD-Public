```python
import pandas as pd
data=pd.read_csv('output6.csv',)
```

```python
data=data[["alpha","gamma","kappa","specificity","accuracy"]]
```

```python
#sort data by specificity
data=data.sort_values(by=['kappa'],ascending=False)
data
```

```python
#plot specificity and kappa for different values of gamma
import matplotlib.pyplot as plt
import numpy as np
data=data.sort_values(by=['specificity'],ascending=False)
x=np.arange(len(data))
plt.plot(x,data["accuracy"])
plt.plot(x,data["specificity"])
plt.xlabel("gamma")
plt.ylabel("kappa")
plt.title("kappa vs gamma")
plt.show()
```

```python
tolerance=0.05
intersection=data[abs(data["specificity"]-data["accuracy"])<tolerance]
intersection
```
